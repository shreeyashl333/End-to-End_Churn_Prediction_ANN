```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
import pickle

pd.set_option('display.max_colwidth', 200)
```

```python
data =  pd.read_csv("Churn_Modelling.csv")
display(data.shape)
display(data.head())
display(data.info())
display(data.describe(include='all'))
df=data.copy()
```

### Some basic preprocessing

```python
df = df.drop(columns=['RowNumber','CustomerId','Surname'], axis=1)
```

```python
display(df.head())
```

```python
le_gender = LabelEncoder()
ohe_geo = OneHotEncoder(sparse_output=False, handle_unknown="ignore")
df['Gender'] = le_gender.fit_transform(df['Gender'])
geo_encoded = ohe_geo.fit_transform(df[["Geography"]])
```

```python
geo_encoded
```

```python
ohe_geo.get_feature_names_out(['Geography'])
```

```python
geo_encoded_df = pd.DataFrame(
    geo_encoded,
    columns=ohe_geo.get_feature_names_out(["Geography"]),
    index=df.index
)

df = df.drop(columns="Geography").join(geo_encoded_df)
```

```python
df.head()
```

```python
with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(le_gender,file)

with open('ohe_geo.pkl','wb') as file:
    pickle.dump(ohe_geo,file)
```

```python
# X and y
X = df.drop('Exited', axis=1)
y = df['Exited']

# train - test 
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.8, random_state=42)

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
with open('scaler.pkl', 'wb') as file:
    pickle.dump(scaler,file)
```

ANN Implementation

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime    
```

```python
(X_train.shape[1],)
```

```python
# Build ANN

model = Sequential([
    Dense(64,activation='relu',input_shape=(X_train.shape[1],)),       # HL1 connected with Input layer
    Dense(32,activation='relu'),                                         # HL2
    Dense(1,activation='sigmoid')                                       # Output Layer
])
model.summary()
```

```python
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss=tensorflow.keras.losses.BinaryCrossentropy()
```

```python
# complie
model.compile(optimizer=opt, loss=loss, metrics=['accuracy'])
```

```python
# tensorboard
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard

log_dir = 'logs/fit/' + datetime.datetime.now().strftime('%Y%m%d-%H%M%S')
tensorflow_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
#  early stop
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
# Train
history = model.fit(
    X_train_scaled, 
    y_train, 
    validation_data = (X_test,y_test),
    epochs = 100,
    callbacks = [tensorflow_callback,early_stopping_callback]  
)
```

```python
model.save('model.h5')
```
